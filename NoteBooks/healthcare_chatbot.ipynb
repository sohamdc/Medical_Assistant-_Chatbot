```python
df.fillna('Not Provided',inplace=True)
df.head()
```

```python
df.dropna(axis=1,inplace=True)
df.info()
```

```python
symp_cols = df[df.columns[1:]].columns
symptoms = []
for col in symp_cols :
    symptoms = symptoms + list(df[symp_cols[0]].values)
```

```python
from collections import Counter
values = Counter(symptoms)
values
```

```python
df.head()
```

```python
df["symptoms"] = df['Symptom_1'] + "--" + df['Symptom_2'] + "--" + df['Symptom_3'] + "--" + df['Symptom_4'] + "--" + df['Symptom_5'] + "--" + \
        df['Symptom_6'] + "--" + df['Symptom_7'] + "--" + df['Symptom_8'] + "--" + df['Symptom_9'] + "--" + df['Symptom_10'] + "--" + \
        df['Symptom_11'] + "--" + df['Symptom_12'] + "--" + df['Symptom_13'] + "--" + df['Symptom_14'] + "--" + df['Symptom_15'] + "--" + \
        df['Symptom_16'] + "--" + df['Symptom_17']

df['symptoms'].head()
```

```python
df = df[["Disease","symptoms"]]
```

```python
df["symptoms"].apply(lambda x: 1 if "itching" in x else 0)
```

```python
results = pd.Series(dict(values))
for item in results.index:
    symptom = item.strip()
    df[f"{symptom}"] = df["symptoms"].apply(lambda x : 1 if symptom in x else 0)

results.index
```

```python
df.drop(["symptoms"],inplace=True,axis=1)
df.head()
```

```python
x = df.iloc[:,1:]
y = df['Disease']
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,shuffle=True)
```

```python
x_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
```

```python
lr=LogisticRegression()
rfc=RandomForestClassifier()
knn = KNeighborsClassifier()
svc = SVC()
models = []
```

```python
models.append(('Logistic Regression',lr))
models.append(('Random Forest',rfc))
models.append(('Knn Classifier',knn))
models.append(('SVM',svc))

algo = []
score = []

for name, model in models:
  model.fit(x_train,y_train)
  y_pred = model.predict(x_test)
  algo.append(name)
  score.append(accuracy_score(y_test,y_pred))

data = {'Algorithm':algo,'Accuracy':score}
algo_compare = pd.DataFrame(data)
algo_compare
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'C': [0.1, 1, 10, 100, 1000], 
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']} 
  
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3)
grid.fit(x_train, y_train)
```

```python
print(grid.best_params_)
print(grid.best_estimator_)
```

```python
model = SVC(C=10, gamma=0.1,kernel='rbf')
model.fit(x_train,y_train)
y_pred = model.predict(x_test)

print(classification_report(y_test,y_pred))
```
